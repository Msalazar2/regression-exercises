## Question 3

Does carat affect price?

Both continuous features will use another scatterplot

```python
sns.relplot(data = train, x = 'carat', y = 'price')
```

Ho = ther is no correlation between carat and price

Ha = there is a correlation between carat and price

```python
r, p = stats.spearmanr(train.carat, train.price)

r, p
```

```python
def spear_test(df, continuous, continuous2):
    
    r, p = stats.spearmanr(df[continuous], df[continuous2])

    print(f'r = {r}')
    print(f'p = {p}')
    
    print('')
    
    a = .05

    if p < a:

        print('Reject the null')

    else:

        print('Fail to reject the null')
```

```python
spear_test(train, 'carat', 'price')
```

There is a strong positive correlation between carat and price. The relationship is significant and carat can potentially help my model predict price.
